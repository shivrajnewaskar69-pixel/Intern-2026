# 1. Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary withword_count, char_count (excluding spaces), unique_words (case-insensitive), and most_frequent_word. Test it with:"The quick brown fox jumps over the lazy dog the fox"

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    freq = {}
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

    most_word = max(freq, key=freq.get)

    return {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words)),
        "most_frequent_word": most_word
    }

sentence = "The quick brown fox jumps over the lazy dog the fox"
print(analyze_sentence(sentence))
```

# 2. Given the string below, do the following and print each result:
text = "A man a plan a canal Panama"
a) Reverse the string without using slicing.
b) Check if it is a palindrome, ignoring case and spaces.
c) Replace every vowel with *.
d) Count how many times each vowel appears.

```python
text = "A man a plan a canal Panama"

# a) Reverse the string without using slicing
reverse_text = ""
for ch in text:
    reverse_text = ch + reverse_text

print("Reversed String:", reverse_text)

# b) Check palindrome (ignore case and spaces)
new_text = text.replace(" ", "").lower()

reverse_new = ""
for ch in new_text:
    reverse_new = ch + reverse_new

if new_text == reverse_new:
    print("Palindrome")
else:
    print("Not a Palindrome")

# c) Replace every vowel with *
result = ""
vowels = "aeiouAEIOU"

for ch in text:
    if ch in vowels:
        result += "*"
    else:
        result += ch

print("Vowels Replaced:", result)

# d) Count how many times each vowel appears
count = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}

for ch in text.lower():
    if ch in count:
        count[ch] += 1

print("Vowel Counts:", count)
```

# 3. Write a function format_report(data) that takes a list of tuples (name, score, grade) and prints aformatted table where name is left-aligned (width 15), score is right-aligned (width 6), and grade iscentered (width 8). Use this data:data = [("Alice", 92, "A"), ("Bob", 78, "B+"), ("Charlie", 55, "C"), ("Diana",100, "A+")]

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
    print("-" * 29)

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```
