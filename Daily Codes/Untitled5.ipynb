# write a program to find the factorial of the number using while loop

```python
num= int(input("enter the number"))
count =1
fact =1
while count<=num:
    fact=fact*count
    count+=1
print(fact)
```

# find the addition of digits of a number
#eg-> make use of % and //

```python
num=12345
add=0
while num>0:
    rem=num%10
    add=add+rem
    num=num//10
print(add)
    
```

# break,continue

```python
for i in(range(1,11)):
    if i==5:
        break
    print(i)
    print("hmm")
```

```python
for i in(range(1,11)):
    print(i)
    print("hmm")
    if i==5:
        break
```

```python
for i in(range(1,11)):
    print(i)
    print("hmm")
    if i==5:
        continue
```
